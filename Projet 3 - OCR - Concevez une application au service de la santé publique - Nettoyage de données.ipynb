### Calcul de données manquantes

```python
def données_manquantes (x):
    données_manquantes = x.isna().sum()
    taux_données_manquantes = round(x.isna().sum()*100/len(x),1)
    tableau_données_manquantes = pd.DataFrame({"nombre de données manquantes":données_manquantes,
                                               "taux de données manquantes":taux_données_manquantes})
    tableau_données_manquantes = tableau_données_manquantes.sort_values("nombre de données manquantes",
                                                                        ascending=False)
    variables_manquantes_50 = tableau_données_manquantes[tableau_données_manquantes["taux de données manquantes"] >= 50]
    print ("{} variables de la dataset présentent un taux de valeurs manquantes supérieur ou égal à 50%".format(len(variables_manquantes_50)))
    return tableau_données_manquantes[:5]
```

```python
pd.set_option("display.max_rows",None)
données_manquantes(data)
```

## Sélection d'observations

## Suppression des valeurs aberrantes/atypiques
- une variable __xxx_100g__ supérieure à 100g (à l'exception de __energy_100g__)
- une variable avec une valeur négative
- une entrée pour la variable `energy_100g` supérieure à 3700kJ (à l'exception de l'huile)
- plus de __sucres__ que de __glucides__

```python
liste_données_aberrantes = (df_france["fat_100g"] > 100) |\
                           (df_france["carbohydrates_100g"] > 100) |\
                           (df_france["proteins_100g"] > 100) |\
                           (df_france["salt_100g"] > 100) |\
                           (df_france["fat_100g"] < 0) |\
                           (df_france["carbohydrates_100g"] < 0) |\
                           (df_france["sugars_100g"] < 0) |\
                           (df_france["proteins_100g"] < 0) |\
                           (df_france["salt_100g"] < 0) |\
                           (df_france["sugars_100g"] > df_france["carbohydrates_100g"]) |\
                           ((df_france["energy_100g"] > 3700) & (df_france["fat_100g"] < 100)) |\
                           (df_france["energy_100g"] > 10000)
df_france = df_france[~liste_données_aberrantes]
df_france.shape
```

- Appliquons la méthode `drop_duplicates()`  à notre dataframe afin de supprimer d'éventuelles doublons exactes 

```python
df_nutri_france = df_france.drop_duplicates().reset_index(drop=True)
```

- Nous allons nettoyer les données textuelles en mettant en miniscule les variables : `product_name` et `countries_en`et en majuscule la variable : `nutriscore_grade`

```python
df_nutri_france["product_name"] = df_nutri_france["product_name"].str.strip().str.lower()
df_nutri_france["countries_en"] = df_nutri_france["countries_en"].str.lower()
df_nutri_france["nutriscore_grade"] = df_nutri_france["nutriscore_grade"].str.capitalize()
```

## Création de variables

- Nous allons rajouter des variables correspondants aux recommendations journalères de la FDA. 
- "Med" : correspond au range de la __dose journalière__ à ne pas dépasser.
- "High" : signifie que la valeur nutritionelle de l'aliment dépasse la dose maximale recommendée par la FDA. 

```python
conditionlist = [(df_nutri_france['salt_100g'] < 2),(df_nutri_france['salt_100g'] > 2) & 
                 (df_nutri_france['salt_100g'] < 5),(df_nutri_france['salt_100g'] >= 5)] 
choicelist = ["Low","Med",'High']
df_nutri_france["salt_indicator"]=np.select(conditionlist, choicelist, default='Not Specified')

conditionlist = [(df_nutri_france['carbohydrates_100g'] < 225),(df_nutri_france['carbohydrates_100g'] > 225) & 
                 (df_nutri_france['carbohydrates_100g'] < 325),(df_nutri_france['carbohydrates_100g'] >= 325)] 
choicelist = ["Low","Med",'High']
df_nutri_france["carbohydrates_indicator"]=np.select(conditionlist, choicelist, default='Not Specified')

conditionlist = [(df_nutri_france['sugars_100g'] < 12),(df_nutri_france['sugars_100g'] > 12) & 
                 (df_nutri_france['sugars_100g'] < 25),(df_nutri_france['sugars_100g'] >= 25)] 
choicelist = ["Low","Med",'High']
df_nutri_france["sugars_indicator"]=np.select(conditionlist, choicelist, default='Not Specified')

conditionlist = [(df_nutri_france['fat_100g'] < 3),(df_nutri_france['fat_100g'] > 3) & 
                 (df_nutri_france['fat_100g'] < 17.5),(df_nutri_france['fat_100g'] >= 17.5)] 
choicelist = ["Low","Med",'High']
df_nutri_france["fat_indicator"]=np.select(conditionlist, choicelist, default='Not Specified')

conditionlist = [(df_nutri_france['proteins_100g'] < 50),(df_nutri_france['proteins_100g'] > 50) & 
                 (df_nutri_france['proteins_100g'] < 70),(df_nutri_france['proteins_100g'] >= 70)] 
choicelist = ["Low","Med",'High']
df_nutri_france["proteins_indicator"]=np.select(conditionlist, choicelist, default='Not Specified')
```

```python
indicators = ["salt_indicator","carbohydrates_indicator","sugars_indicator","fat_indicator",
              "proteins_indicator"]
print(df_nutri_france.shape)
df_nutri_france[indicators].head(3)
```
